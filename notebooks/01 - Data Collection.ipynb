```python
import pandas as pd
import numpy as np

# remove column restrictions
pd.options.display.max_columns = None
```

```python
# scan raw data from ../data/raw
```

```python
raw_data_path = '../data/raw'

# List all files in the raw data directory
import os
raw_files = os.listdir(raw_data_path)
print("Raw data files:", raw_files)
```

```python
for file in raw_files:
    file_path = os.path.join(raw_data_path, file)
    df = pd.read_csv(file_path, encoding="latin-1", engine="python")
    print(f"Data from {file}:")
    display(df.head())  # Display the first few rows of each file
    print("\n")
    print(f"Shape of {file}: {df.shape}")
```

```python
# get a sample of this raw trade data
sample_data = df.sample(n=5, random_state=42)
display(sample_data)
```

```python
# save to csv
sample_data.to_csv('../data/processed/sample_trade_data.csv', index=False)
```
